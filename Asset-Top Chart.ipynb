```python
import pandas as pd
import matplotlib.pyplot as plt

# Declaration
limit = 7
unit = "Cal"
data = pd.read_csv("Kumande Asset - Food.csv")
msg = f"Top {limit} Foods by Calorie Count\n"

# Sorting
sorted_data = data.sort_values(by="calorie", ascending=False)
top = sorted_data[["food", "calorie"]].head(limit)
result = list(top.itertuples(index=False, name=None))

# List
print(msg)
for idx, (food, calorie) in enumerate(result):
    print(f"{idx + 1}. {food} - {calorie} {unit}")
print("\n")

# Dataset
context = [item[0] for item in result] 
total = [item[1] for item in result]  

# Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(context, total, color="skyblue")

plt.title(msg.strip()) 
plt.xlabel("Food")
plt.ylabel(f"Calorie Count ({unit})")
plt.xticks(rotation=45) 

plt.tight_layout() 
plt.show()
```
