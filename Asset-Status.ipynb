```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Declaration
data = pd.read_csv("Kumande Asset - Food.csv")

# Helpers
def calorie_status(calorie): # Per Serving
    if calorie > 250:
        return "High"
    elif calorie < 100:
        return "Low"
    else:
        return "Normal"

# Sorting
filtered_data = data[["food", "calorie"]].copy()
filtered_data = filtered_data.sort_values(by="food")

# Table Prepare
filtered_data["food"] = filtered_data["food"].str.title()
filtered_data["Status"] = filtered_data["calorie"].apply(calorie_status)
filtered_data.rename(columns={"food": "Food","calorie": "Calories (Cal)"}, inplace=True)

# Dataset
status = dict(Counter(filtered_data["Status"]))
context = list(status.keys()) 
total = list(status.values()) 

# Pie Chart
plt.figure(figsize=(8,8))
plt.pie(total,labels=context,autopct="%1.1f%%",startangle=140)
plt.title("Calorie Status Distribution")
plt.show()

# Table Init
df = filtered_data.style.set_properties(**{"text-align":"left"})
df.set_caption("Calorie Status")
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Declaration 
data = pd.read_csv("Kumande Asset - Food.csv")

# Helpers 
def fat_status(fat): # Per serving
    if fat > 7:
        return "High"
    elif fat < 3:
        return "Low"
    else:
        return "Normal"

# Sorting
filtered_data = data[["food","Fat"]].copy()
filtered_data = filtered_data.sort_values(by="food")

# Table Prepare
filtered_data["food"] = filtered_data["food"].str.title()
filtered_data["Status"] = filtered_data["Fat"].apply(fat_status)
filtered_data.rename(columns={"food":"Food","Fat":"Fat (g)"},inplace=True)

# Dataset
status = dict(Counter(filtered_data["Status"]))
context =  list(status.keys())
total = list(status.values())

# Pie Chart
plt.figure(figsize=(8,8))
plt.pie(total,labels=context,autopct="%1.1f%%",startangle=140)
plt.title("Fat Status Distribution")
plt.show()

# Table Init
df = filtered_data.style.format({"Fat (g)": "{:.2f}"}).set_properties(**{"text-align":"left"})
df.set_caption("Fat Status")
df
```
